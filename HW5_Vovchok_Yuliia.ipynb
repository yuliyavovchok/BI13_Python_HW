```python
from itertools import chain, combinations, permutations
from collections import defaultdict, Counter
import sys
import os
```

```python
with open('orders.txt', 'r') as file:
    data = file.read().split('\n\n')
```

```python
# Генеруємо лист з ордерів
orders = [order.split('@@@') for order in data]

#print(orders)
print('Total orders: ', len(orders))
print(len(orders[0]))
```

```python
# Створюємо словник для підрахунку к-сті кожного продукту
prod_count = Counter(chain(*orders))

print(len(prod_count))
```

```python
sum(prod_count.values())
```

```python
print(len(list(chain(*orders))))
print(len(set(chain(*orders))))
```

```python
# Створюємо список усіх пар продуктів у чеках (порядок має значення)
all_pairs = []
for i in orders:
    c = permutations(i, 2)
    all_pairs.extend(c)
```

```python
print(len(all_pairs))
```

```python
# Створюємо словник для підрахунку к-сті пар продуктів
prod_pair_count = Counter(all_pairs)

print(len(prod_pair_count))
```

```python
sum(prod_pair_count.values())
```

```python
prod_pair_count
```

```python
# Створюємо функцію для знаходження всіх пар продуктів, які будуть задовольняти умову впевненості та підтримки
def print_associations(min_conf, min_support):
    for (x, y) in prod_pair_count.keys():
        conf = round(prod_pair_count[(x, y)]/prod_count[x], 4)
        support = prod_pair_count[(x, y)]
        if conf >= min_conf and support >= min_support:
            print(f'{x} => {y} ({conf * 100}% confidence), {support} support')
```

```python
print_associations(0.2, 1312)
```
