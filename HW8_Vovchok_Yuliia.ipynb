```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import requests
```

```python
URL = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-07-07/coffee_ratings.csv'

req = requests.get(URL)
```

```python
match req.status_code:
    case 200: 
        df = pd.read_csv(URL)
    case 404: 
        raise 'Problem with connection'
    case _ :
        raise f'Something wrong {req.status_code}'
```

2. Які кореляції між різними показниками оцінки кави?

```python
df.describe()
```

```python
df[['total_cup_points', 'aroma', 'flavor', 'aftertaste', 'acidity', 'body', 'balance', 'uniformity', 'clean_cup',  'sweetness', 'cupper_points', 'moisture', 'category_one_defects', 'quakers', 'category_two_defects']].corr()
```

```python
# Найбільш корелюють з загальною оцінкою: 'flavor', 'aftertaste', 'balance' і 'acidity'.
# 'moisture' - не корелює з жодним показником.
# Відносно мала кореляція з іншими показниками: 'uniformity', 'clean_cup' і 'sweetness'. Хоча з 'total_cup_points' кореляція середня, але найнижча відносно інших показників
corr1=df[['total_cup_points', 'aroma', 'flavor', 'aftertaste', 'acidity', 'body', 'balance', 'uniformity', 'clean_cup',  'sweetness', 'cupper_points', 'moisture']].corr()
sns.heatmap(corr1, annot=True)
```

```python
#Бачимо, що різні оцінки дефектів кави ймовірно малокорелюють з загальною оцінкою
corr2=df[['total_cup_points', 'category_one_defects', 'quakers', 'category_two_defects']].corr()
sns.heatmap(corr2, annot=True)
```

```python
# Графік кореляції показників з найбільшою взаємозалежністю.
# Найбільшкорелюють пари: 'flavor' і 'aftertaste', 'total_cup_points' і 'flavor', 'total_cup_points' і 'aftertaste'.
corr3=df[['total_cup_points', 'aroma', 'flavor', 'aftertaste', 'acidity', 'body', 'balance', 'cupper_points']].corr()
sns.heatmap(corr3, annot=True)
```

```python
df2 = df[['species', 'total_cup_points', 'aroma', 'flavor', 'aftertaste', 'acidity', 'body', 'balance', 'uniformity', 'clean_cup',  'sweetness', 'cupper_points', 'moisture']]
sns.pairplot(df2, hue='species')
```

4. Чи впливає країна походження на якість кави?

```python
plt.subplots(figsize = (8,6))
sns.boxplot(data = df3, y = 'country_of_origin', x = 'total_cup_points');
```

```python
sns.barplot(x='total_cup_points', y='country_of_origin', data= df3)
plt.title("Середній показник оцінок по країні походження")
plt.axvline(x=82, c="green", label="y=0")
plt.axvline(x=86, c="blue", label="y=0")
plt.axvline(x=77, c="red", label="y=0")
plt.show()
```

```python
df4 = df3.groupby('country_of_origin')['total_cup_points'].mean().sort_values(ascending=False)
df4
```

```python
pivot4=df3.pivot_table(index=["country_of_origin"],values=['total_cup_points'])
plt.subplots(figsize = (5,8))
sns.heatmap(pivot4)
plt.show()
```

Висновок: На верхній графік додано три вертикальні лінії, які відображають мінімальний, середній та найвищий показники оцінок. Нижній графік - теплова карта. Ми бачимо, що:
- Найбільшу середню оцінку отримала кава з Papua New Guinea та Ethiopia (понад 85 балів в середньому)
- Найнижча середня оцінка отримала кава з Haiti
Загалом діапазон середньої оцінки коливається від 77 до 86 балів, і країни дуже щільно знаходяться одна біля одної. Тому можна сказати, що хоч країна походження кави впливає на оцінку, але цей вплив не є надто суттєвим.
